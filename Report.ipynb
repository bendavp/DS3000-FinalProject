###Significance of the Problem:

The DeepWater Horizon spill was not the first major oil spill, nor will it be the last. Oil spills can have far reaching effects on ecosystems. This project is important because it can provide insights into how shore birds are affected by oil spills and the physiological markers. Dramatic changes in physiology can have long term effects, such as affecting fecundity, so the results of this model can help predict how shore birds are impacted by an oil spill. It can also be used to see how far the impacts of an oil spill can be seen. Since many birds migrate to or through the Gulf of Mexico, many more birds may have been impacted by the spill than initially thought. Therefore, finding a simple way to test a bird for oil exposure is imperative to discovering how far the effects of the oil spill traveled. 

The external effects of oil have been vastly researched, showing how oil 
removes the water repellent properties of feathers (Morandin and O’Hara 2016). This has an effect on buoyancy, can lead to behavioral changes, and can cause impaired flight. Exposure and ingestion of oil have been shown to have metabolic effects on marine birds, when not in lethal doses (Morandin and O’Hara 2016). Birds exposed to oil have been shown to be less able to cope with other stressors than birds not exposed to oil (Lattin et al. 2014). Polycyclic aromatic hydrocarbons (PAHs) are a chemical compound that are derived from fossil fuels. These PAHs are toxic at low concentrations and can be transferred up the food chain. Even when this compound is not consumed in lethal doses, it can have serious physiological effects like liver damage, anemia, immune suppression, gastrointestinal damage etc. (Paruk et al. 2016). Seabirds may be more at risk of the harmful effects than terrestrial birds of oil due to their longer time spent maturing and the fact that many species only lay one egg per mating season (Morandin and O’hara 2016). This means the species would be less able to adapt and evolve to these harmful conditions. The effects of oil pollution may also cause lasting effects in the reproductive ability of birds exposed to oil, so the population of shorebirds may be impacted for more than one generation.

We are seeing if we can accurately predict if a bird was exposed to oil pollution. A better understanding of physiological markers in shore birds helps researchers to target certain markers to classify whether a bird has been involved in an oil spill; this can help identify unknown oil spills.

````python
# drop all rows where all columns from "Heinz Body Count" to the end of the DataFrame
# are None values (includes "NR", "ND", and " ")
columns_to_analyze = list(df.columns[14:])
clean_df = df.replace(" ", np.NaN).replace("NR", np.NaN).replace("ND", np.NaN).dropna(how="all", subset=columns_to_analyze).reset_index().drop("index", axis=1)
clean_df
````

````python
# dropping the SSSP species as there is no usable data
# starting from the "Heinz Body Count" column, so there
# is not enough data to give to a model such that it would
# be able to reliably predict that a bird in this species
# is a reference or impacted
clean_df = clean_df[clean_df["Species"] != "SSSP"]
````

````python
# for Heinz Body Count, missing values are replaced with 0
# because background research shows that this value is typically 0
# and most of the other values in this column were 0

clean_df["Heinz Body Count"] = clean_df["Heinz Body Count"].apply(lambda x: 0 if math.isnan(x) else x)
````

````python
# Functions for cleaning up data

# aggregrates the median for a given column by species and classification; 
# replaces the nan values in the given column by the median based on the 
# species and classification
def replaceNanWithMedian(column, df_to_clean): 
  res_df = df_to_clean.copy()
  clean_df_grouped = df_to_clean.groupby(["Species", "Classification"])
  medians = clean_df_grouped[column].median().to_dict()
  for index, row in df_to_clean.iterrows():
    if math.isnan(row[column]):
      res_df.at[index, column] = medians.get((row["Species"], row["Classification"]))
  return res_df

# drops the given column from the dataframe if there is an 
# aggregrated median of nan for the species and classification
def dropUnusableCols(column, df_to_clean):
  res_df = df_to_clean.copy()
  clean_df_grouped = df_to_clean.groupby(["Species", "Classification"])
  medians = clean_df_grouped[column].median().to_dict()
  for key in medians.keys():
    if math.isnan(medians.get(key)):
      res_df = res_df.drop(columns=column)
      break
  return res_df
````

````python
# Here, we are cleaning continous data in the dataframe:

# categorical columns that we do not want to clean, as right now we are
# only dealing with continuous data
cat_columns = ["Band Number", "Date", "Capture Site", "State", "Latitude", "Longitude", "Site Characteristics", "Classification", 
                     "Species", "Age", "Bird Weight (g)", "Visible Oil Assessment", "UV Light Oil Assessment", "Oil at Capture Site", 
                     "Heinz Body Count", "Buffy Coat", "Polychromasia", 	"Anisocytosis", "RBC Count (million/cubic mm)","Serum Condition"]

# creating list of columns that are continous, based on the categorical
# columns above
columns_median_fix = clean_df.columns.to_list() 
for col in cat_columns:
  columns_median_fix.remove(col)

# replaces the NaN in the columns with continuous data with the median,
# based on function in above cell 
new_ = clean_df.copy()
for feature in columns_median_fix:
  new_ = replaceNanWithMedian(feature, new_)

# removing columns where any aggregrate median is nan
for feature in columns_median_fix:
  new_ = dropUnusableCols(feature, new_)

clean_df = new_
````

````python
# Cleaning up categorical data in the DataFrame:

# removing Site Characteristics because it is not possible to clean 
# this column since we can only fill NaN with Unknowns 
# (doesn't make sense in the long run)
clean_df.drop("Site Characteristics", axis=1, inplace=True)

# Removing Age and Bird Weight (g)
# Age doesn't appear to be useful for determining if the bird is from an impacted site
# Bird Weight had some clearly incorrect values, and fixing them is not possible
fig = px.scatter(clean_df, x="Age", y="Bird Weight (g)", color="Species")
fig.show()
clean_df.drop("Age", axis=1, inplace=True)
clean_df.drop("Bird Weight (g)", axis=1, inplace=True)

# removing Visible Oil Assessment
# removing UV Light Oil Assessment
# removing Oil at Capture Site
# These would not give in depth physiological evidence for whether the 
# bird was in an oil spill. The current values would likely lead to our model
# classifying all records with any value other than "None" as impacted, when
# our goal is to identify physiological markers of impacted birds, rather
# than physical markers.
clean_df.groupby(["Species", "Classification"])["Visible Oil Assessment"].describe()
clean_df.groupby(["Species", "Classification"])["UV Light Oil Assessment"].describe()
clean_df.groupby(["Species", "Classification"])["Oil at Capture Site"].describe()
clean_df.drop("Visible Oil Assessment", axis=1, inplace=True)
clean_df.drop("UV Light Oil Assessment", axis=1, inplace=True)
clean_df.drop("Oil at Capture Site", axis=1, inplace=True)

# removing RBC Count (million/cubic mm)
# Unable to resolve NaNs and combine both discrete and continous data that
# is found in the column.
clean_df.groupby(["Species", "Classification"])["RBC Count (million/cubic mm)"].describe()
clean_df.drop("RBC Count (million/cubic mm)", axis=1, inplace=True)

# removing Polychromasia
# The top frequency for all records is slight, meaning this column is unlikely
# to contain data useful to our model.
clean_df.groupby(["Species", "Classification"])["Polychromasia"].describe()
clean_df.drop("Polychromasia", axis=1, inplace=True)

# removing Anisocytosis
# The top frequency for all records is slight, meaning this column is unlikely
# to contain data useful to our model.
clean_df.groupby(["Species", "Classification"])["Anisocytosis"].describe()
clean_df.drop("Anisocytosis", axis=1, inplace=True)

# removing Serum Condition
# This column is simply a "general description" String, which cannot
# be normalized into discrete bins. It may be possible to perform
# sentiment analysis on these Strings to give our classification model,
# however this seems beyond the scope of this project.
clean_df.groupby(["Species", "Classification"])["Serum Condition"].describe()
clean_df.drop("Serum Condition", axis=1, inplace=True)

# removing Buffy Coats
# The data is sparse, therefore we cannot pull enough useful information out
# of this column. It is likely that the researchers could not obtain
# enough whole blood for proper analysis
clean_df["Buffy Coat"].isna().sum()
clean_df.drop("Buffy Coat", axis=1, inplace=True)

# removing Date
# This column doesn't provide any relevant physiological data
clean_df.drop("Date", axis=1, inplace=True)

# removing Capture Site 
# Can get the location (Capture Site data) via latitude and longitude, so 
# this String does not help us
clean_df.drop("Capture Site", axis=1, inplace=True)

# removing State 
# Can get the State data via latitude and longitude
clean_df.drop("State", axis=1, inplace=True)

# There are no more NaN values in our cleaned dataframe
clean_df.isna().sum()

# Dropping band number as it is just a key 
# and is not useful for actual prediction
clean_df.drop("Band Number", axis=1, inplace=True)

# Dropping longitude and latitude because they are not physiological markers,
# and we do not want to overfit the model to our particular geographical data
clean_df.drop("Longitude", axis=1, inplace=True)
clean_df.drop("Latitude", axis=1, inplace=True)
````

Next, we define all the classifiers we will use (since we have a classification project), and we will use both cross validation (KFold) and `train_test_split` to split our _scaled_ features, then train the algorithms on the _scaled_ features, then evaluate metrics in those algorithms. We will also hypertune our parameters in each model to avoid overfitting/underfitting to our data.


````python
classifiers = {
    "kNN" : KNeighborsClassifier(n_neighbors=3),
    "SVM" : LinearSVC(max_iter=100000, random_state=42069),
    "GaussianNB" : GaussianNB(),
    "Tree" : DecisionTreeClassifier(random_state=42069),
    "Logic" : LogisticRegression(random_state=42069)
}
````

````python
# define MinMaxScaler
scaler = MinMaxScaler()

# scale features first, before we use multiple model selection
scaler.fit(features)
features_scaled = scaler.transform(features)

# KFold for cross validation
kfold = KFold(n_splits=10, random_state=42069, shuffle=True)

# test all above models to see which one is best for our data
for classifier_name, classifier_object in classifiers.items():
  # array of correct predictions over total predictions for each fold 
  scores = cross_val_score(estimator=classifier_object, X=features_scaled, y=target, cv=kfold)


  print(classifier_name + ": \n\t" 
        # mean of distribution of accuracies by fold
        + f'mean accuracy={scores.mean():.2%} ' 
        # standard deviation of distribution of fold accuracies
        + f'standard deviation={scores.std():.2%}' + "\n")

# split data into training and testing set
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=42069)

# scale training and testing set
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

for classifier_name, classifier_object in classifiers.items():
  
  classifier_object.fit(X=X_train_scaled, y=y_train)

  predicted = classifier_object.predict(X=X_test_scaled)

  # percent of correct predictions over total predictions
  accuracy = classifier_object.score(X_test_scaled, y_test)
  print(classifier_name + ": \n\t" + f'Predicted accuracy on the test data: {accuracy*100:.2f}%' + "\n")
````

Based on these results, we will use the Decision Tree Classifier, since it
gave us the best accuracy and a relatively small and consistent standard deviation with and without hyperparameter tuning. We considered using kNN, as it too had a high accuracy and low standard deviation, although this result was only after tuning n_neighbors to 3, which increases the risk of underfitting the model to our data. To prevent this risk, we will stick to Decision Trees, as this gave us the most reliable accuracy and standard deviation, even when hyperparameter tuning max_depth. We also attempted to hypertune the other classifiers, but none gave more accurate results compared to the defaults.

Now we will perform our hypothesis tests:

````python
# get a list of unique species
unique_species = []
for bird in species[:-1]:
  unique_species.append(bird[0])
unique_species=list(set(unique_species))

# function to run an independent samples t-test to determine if 
# a feature is significantly different among reference and impacted 
# groups of all species
def t_test_by_species(df_, column_):
  for bird in unique_species:
    species_df = df_[df_["Species"] == bird]
    reference = species_df[species_df["Classification"] == "Reference"][column_]
    impact = species_df[species_df["Classification"] == "Impacted"][column_]
    t_test = stats.ttest_ind(reference, impact)
    deg_f = len(reference) + len(impact) - 2
    print(f'Species: \"{bird}\"')
    print(f'\tdf={deg_f}')
    print(f'\ttest statistic={t_test.statistic}')
    print(f'\tp-value={t_test.pvalue}')
    print(f'\tReference Stats:')
    print(f'\t\tN={len(reference)}')
    print(f'\t\tmean={reference.mean()}')
    print(f'\t\tstd err={reference.sem()}')
    print(f'\tImpacted Stats:')
    print(f'\t\tN={len(impact)}')
    print(f'\t\tmean={impact.mean()}')
    print(f'\t\tstd err={impact.sem()}')
    if t_test.pvalue <= 0.05:
      print(f'These results indicate that the {"Impacted" if impact.mean() > reference.mean() else "Reference"} group has a significantly higher {column_} than the {"Reference" if impact.mean() > reference.mean() else "Impacted"} group.')
    else: 
      print(f'These results indicate that there is no significant difference between the {column_} means of the Impacted and Reference groups.')
    print()
````

When running the t-test, it is clear that 3 out of the 5 species do not support our hypothesis, as their PAH values are not significantly different from each other. Therefore, we fail to reject our null hypothesis, which is that the PAH levels between reference and impacted birds are not different.

## 4. DISCUSSION


In Section 3.3, we ran independent sample t-tests was conducted to compare the differences in PAH (ng/mL), Reticulocyte Count, and Uric Acid (mg/dl) between the impacted and reference groups. These t-tests were run separately for each species. 

Here, the null hypothesis is that there is no difference between the means of the markers for the impacted and reference groups. The alternative hypothesis is that the markers' means are different between the impacted and reference groups.

A p-value calculated from the t-test less than and equal to 0.05 would signify that there is a significant difference between the two means, and that we will reject our null hypothesis. If the p-value is greater than 0.05, then we fail to reject our null hypothesis because the means are insignificantly different.

As stated above in Section 2, this is a supervised learning classification problem, and we approached data analysis with a hypothesis that higher values of certain physiological markers that indicate contact with oil (PAH (nh/ml), Heinz Body Counts, Reticulocytes Counts, etc.) would correspond with a higher likelihood that the given bird was impacted by an oil spill. Our data analysis also included a lot of manual analysis during the data wrangling and data exploration steps in section 3 when we were determining which features were reasonable for us to keep from the original dataset and which were not.

  After finalizing feature selection in section 3.3, we moved on to evaluate the 5 models described in section 2.3:


*   K-Nearest Neighbors-based classification (KNeighborsClassifier)
*   Linear Support Vector Classification (LinearSVC)
*   Gaussian Naive Bayes (GaussianNB)
*   Decision Tree classifier (DecisionTreeClassifier)
*   Logistic Regression classifier (LogisticRegression)


  We used the exact same method for comparing model performance in section 3.4 as we did in section 3.3: using the aforementioned models, we iteratively tested each one using both cross validation with KFold and a standard 75/25 train-test split. The metric used to evaluate the models was the prediction accuracy on the test data for train-test split, and the average of prediction accuracies for the cross validation. We also performed hyperparameter tuning in this section to determine whether or not any of the models could benefit from it, and we did discover that the K-Nearest Neighbors classifier benefited most from a value of n_neighbors=4, however even after this change the Decision Tree classifier was still our most accurate model.

  Once we selected the Decision Tree classifier as the final candidate for our predictive model we generated a confusion matrix and a classification report using the model’s predictions. The results of the classification report and the confusion matrix indicated that our model had a final accuracy rate of roughly 92% on the testing data, with precision/recall scores for positive and negative classifications being 94%/91% for positive and 90%/92% for negative respectively.
